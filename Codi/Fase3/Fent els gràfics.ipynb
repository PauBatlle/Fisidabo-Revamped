```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
sns.set_style('white')
```

```python
cart = [pd.read_csv("Dades_takes/"+str(i)+"Cart.csv") for i in range(1,5)] #Llegeixo fitxers
cart = [c.drop('Unnamed: 0', 1) for c in cart] #Una guarrada, trec a mà una columna que em sobra
pol = [pd.read_csv("Dades_takes/"+str(i)+"Pol.csv") for i in range(1,5)]
pol = [p.drop('Unnamed: 0', 1) for p in pol] #Una guarrada, trec a mà una columna que em sobra
```

```python
c1,c2,c3,c4 = cart[0],cart[1],cart[2],cart[3]
p1,p2,p3,p4 = pol[0],pol[1],pol[2],pol[3]
```

# Radis

```python
plt.subplot(221)
plt.plot(p1['t'],p1['r'])
plt.title("Take 1")
plt.ylabel("Radius (m)")
plt.xlabel("Time (s)")
plt.subplot(222)
plt.plot(p2['t'],p2['r'])
plt.title("Take 2")
plt.ylabel("Radius (m)")
plt.xlabel("Time (s)")
plt.subplot(223)
plt.plot(p3['t'],p3['r'])
plt.title("Take 3")
plt.ylabel("Radius (m)")
plt.xlabel("Time (s)")
plt.subplot(224)
plt.plot(p4['t'],p4['r'])
plt.title("Take 4")
plt.ylabel("Radius (m)")
plt.xlabel("Time (s)")
plt.subplots_adjust(wspace = .5, hspace=.5)
```

```python
circle1 = plt.Circle((0, 0), 0.216/2, color='r', fill = True)
fig, ax = plt.subplots()
plt.plot(c1['x'],c1['y'])
ax.add_artist(circle1)
plt.ylabel("y (m)")
plt.xlabel("x (m)")
plt.title("Take 1")
plt.axis("equal")
```
