```python
import pandas as pd
from ebola2retall import *
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
%matplotlib


#retorna la distancia quadrat normalitzada de les acceleracions
def distancia_acc (tableau, plot = 0):
    difx = pd.Series.as_matrix(tableau['ax']-tableau['axe'])
    dify = pd.Series.as_matrix(tableau['ay']-tableau['aye'])
    if (plot == True):
        plt.plot(tableau['t'], difx, label = 'x')
        plt.plot(tableau['t'], dify, label = 'y')
        plt.legend()
    return (np.dot(difx,difx) + np.dot(dify,dify))/len(tableau['t'])

def complete_show (table):
    fig = plt.figure('pos-accel-time')
    axe = fig.gca(projection='3d')
    axe.plot(tableau['x'],tableau['y'],tableau['t'], label='trajectòria')
    axe.plot(tableau['ax'],tableau['ay'],tableau['t'], label='acceleració')
    axe.plot(tableau['axe'],tableau['aye'],tableau['t'], label='accel_èbola')
    plt.xlabel('x var')
    plt.ylabel('y var')
    axe.set_zlabel('time')

    axe.legend()
    plt.show()
```

```python
taula1 = pd.read_csv("Dades_polinomis/take1_hr.csv")
taula2 = pd.read_csv("Dades_polinomis/take2_hr.csv")
taula3 = pd.read_csv("Dades_polinomis/take3_hr.csv")
```

```python
factor = 0.3
inicial = np.array([1.68, 0.18, 1.8, 0.0055])
print (inicial)
Delta = inicial * factor

dist_fita = 1
dist_min = [1, None]
for c in np.linspace(inicial[0] - Delta[0]*1.1, inicial[0] + Delta[0]*1.1, 10):
    for k in np.linspace(inicial[1] - Delta[1]*1.1, inicial[1] + Delta[1]*1.1, 10):
        for sigma in np.linspace(inicial[2] - Delta[2], inicial[2] + Delta[2]*1.2, 16):
            for mu in np.linspace(inicial[3] - Delta[3], inicial[3] + Delta[3], 6):
                values = [c,k,sigma, mu]
                tableau1 = calcula(taula1, *values)
                tableau2 = calcula(taula2, *values)
                tableau3 = calcula(taula3, *values)
                
                dist_total = distancia_acc(tableau1)+distancia_acc(tableau2)+distancia_acc(tableau3)
                if (dist_total < dist_fita):
                    print (dist_total, values)
                    if (dist_total < dist_min[0]):
                        dist_min = [dist_total, values]
                        print ("nou minim" + str(dist_min))
```

```python
values = [2.2344, 0.147, 1.26, 0.00583]
values = dist_min[1]
tableau = calcula(taula3, *values)
print (distancia_acc(tableau))
complete_show(tableau)
```
