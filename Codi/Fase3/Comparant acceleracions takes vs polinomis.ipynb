```python
import pandas as pd
from os import listdir
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
polinomis = [pd.read_csv("Dades_polinomis/take"+str(i)+"_hr.csv") for i in range(1,5)]
takes = [pd.read_csv("Dades_takes/"+str(i)+"Cart.csv") for i in range(1,5)]
```

```python
for i in range(1,5):
    t0 = polinomis[i-1]["t"][0]
    s = int(takes[i-1][takes[i-1]["t"] == t0]["Unnamed: 0"])
    takes[i-1] = takes[i-1][s:-1]
    takes[i-1].index = range(len(takes[i-1]))
```

```python
vec = ["x","y"]
for i in range(1,5):
    for z in range(0,2):
        plt.close()
        plt.title("take"+str(i)+vec[z])
        plt.plot(polinomis[i-1][vec[z]]);
        plt.plot(takes[i-1][vec[z]]);
        plt.savefig("take"+str(i)+vec[z]+".png");
```
