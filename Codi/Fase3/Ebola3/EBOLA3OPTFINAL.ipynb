```python
from funcionsopt3 import *
import pandas as pd
import numpy as np
from random import random
from scipy.optimize import minimize
from numpy.linalg import norm
import time
import matplotlib.pyplot as plt
datasets = [pd.read_csv("take"+str(i)+"_hr.csv") for i in range(1,5)]
```

```python
lims = [1,2.5,0.1,0.4,1.2,2,0.002,0.015,0,1,-5,10,-5,10,-5,10]
```

En realitat s'ha d'optimitzar sobre 8 parÃ metres: ws take1, ws take2, ws take 3

```python
def random_dins(limits):
    c = [random(), random(), random(), random(), random(), random(), random(), random()]
    vec = [limits[2*i]*c[i]+limits[2*i+1]*(1-c[i]) for i in range(len(c))]
    return vec
```

```python
def fitness(v, dades_take):
    return distancia_acc(calcula3(dades_take, *v))
def fitness_avg(v,dades_take1 = datasets[0], dades_take2 = datasets[1], dades_take3 = datasets[2]):
    return (fitness(np.concatenate((v[0:5],[v[5]])),dades_take1) + fitness(np.concatenate((v[0:5],[v[6]])),dades_take2) + fitness(np.concatenate((v[0:5],[v[7]])),dades_take3))*(1/3)
```

```python
dades = []
for i in range(nombre_de_dades_a_agafar):
    t0 = time.clock()
    xinicial = random_dins(lims)
    c = minimize(fitness_avg, xinicial, method='Nelder-Mead', options={'disp': True, 'maxiter': 1000,'maxfev': 1000})
    t1 = time.clock()
    print(i,t1-t0)
    dades.append([*xinicial,*c.x,c.fun,t1-t0])
```
